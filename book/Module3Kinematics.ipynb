b) $𝑣\frac {\partial 𝑢}{\partial 𝑦}$

**Solution**

```{dropdown}
This is the local acceleration in the x direction, representing the time rate of change of the velocity at a given location within the flow. It would be zero in a steady flow
```

**Question 1 (b)**

**Solution**



```{dropdown} Hint

**Steady vs Unsteady Flow:**

The flow is **steady** if the partial derivatives with respect to time are zero:
- $ \frac{\partial u}{\partial t} = 0 $
- $ \frac{\partial v}{\partial t} = 0$
- $ \frac{\partial w}{\partial t} = 0 $

If any of these are not zero, the flow is **unsteady**.

**Uniform vs Non-uniform Flow:**

The flow is **uniform** if the velocity does **not vary with space**, meaning:
- $\frac{\partial u}{\partial x} = 0 $
- $ \frac{\partial v}{\partial y} = 0 $
- $ \frac{\partial w}{\partial z} = 0 $

If any of these spatial derivatives are not zero, the flow is **non-uniform**.

```

**Solution**

```{dropdown} Show code cell outputs
The acceleration is calculated with the following equations 

$$𝑎_𝑥 = \frac {\partial 𝑢}{\partial 𝑡} + 𝑢\frac {\partial 𝑢}{\partial 𝑥} + 𝑣\frac {\partial 𝑢}{\partial 𝑦}$$

For $a_x$ the partial differences are: 

  $\frac {\partial 𝑢}{\partial 𝑡} = 12t$ 

  $\frac {\partial 𝑢}{\partial 𝑥}= 0$

  $\frac {\partial 𝑢}{\partial 𝑦} = 0$

Replacing these partial differences in the acceleration equation we have the following:

  $$𝑎_𝑥= (6𝑡^2 + 5)12t + 0 + 0$$
  
Considering this equation for acceleration, the flow is unsteady because there is a local component of the acceleration and is uniform because there is no convective acceleration
```

A viscous fluid flows through a channel with a square cross-section (width and height of 0.5 𝑚) with a vertical velocity profile of 𝑢(𝑧) = 8𝑧<sup>2</sup> , where 𝑢 is the horizontal velocity and 𝑧 is the vertical coordinate. Determine the discharge in the channel.

```{dropdown} Show code cell outputs

$$Q= \int u(z)dA$$

Where:

$u(z)= 8z^2$

$dA= bdz$

The integration is carried out from the bottom of the channel to a height of 0.5 m

$$
\begin{aligned}
\quad Q &= \int_0^h 8z^2 b \, dz \\
&= 8b \int_0^h z^2 \, dz \\
&= 8b \left[ \frac{z^3}{3} \right]_0^h \\
&= 8b \cdot \frac{h^3}{3} \\
\text{Substitute } b = 0.5, \, h = 0.5: \quad Q &= 8 \cdot 0.5 \cdot \frac{(0.5)^3}{3} \\
&= 4 \cdot \frac{0.125}{3} \\
&= \frac{0.5}{3} \\
&\approx 0.167 \, \text{m}^3/\text{s}
\end{aligned}
$$
```
