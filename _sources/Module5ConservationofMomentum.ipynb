```{dropdown} Show code cell outputs

Continuity equation:

\begin{align*}
Q_1 &= Q_2 \\
v_1 A_1 &= v_2 A_2 \\
v_1(y_1 b) &= v_2(y_2 b) \\
v_1 \frac{y_1}{y_2} &= v_2 \\
\end{align*}

Conservation of energy:

\begin{align*}
z_1 + \frac{p_1}{\rho g} + \frac{v_1^2}{2g} &= z_2 + \frac{p_2}{\rho g} + \frac{v_2^2}{2g} + h_f \\
\frac{p_1}{\rho g} &= y_1, \quad \frac{p_2}{\rho g} = y_2 \\
z_1 = z_2 \\
y_1 + \frac{v_1^2}{2g} &= y_2 + \frac{y_1^2 v_1^2}{2g y_2^2} + h_f
\end{align*}

Solving for $( v_1)$:

\begin{align*}
v_1 = \left( \frac{2g(y_2 - y_1 + h_f)}{1 - \frac {y_1^1}{y_2^2}} \right)^{0.5}
\end{align*}

$N \mathrm{II}$:

\begin{align*}
\sum F_x &= \rho Q (v_{2x} - v_{1x}) \\
F_1 - F_2 - R_x &= \rho Q (v_{2x} - v_{1x})
\end{align*}


```

````python
%%render long 
h_f=0.2 #m
y_1=2.2 #m
y_2=0.4 #m
y_3=0.5 #m
b=1 #m


v_1= ((2*g*(y_2-y_1+h_f))/(1-((y_1**2)/(y_2**2))))**0.5 #m/s
v_2= 5.5*v_1 #m/s
A_1= y_1*b #m2
A_2= y_2*b #m2
Q= v_1*A_1 #m3/s
R_x= rho_water*g*y_1/2*y_1-rho_water*g*y_2/2*y_2-rho_water*Q*(v_2-v_1) #N
R_x_=R_x/1000 #kN
````

```{dropdown} Show code cell outputs
The force on the structure is the same size but in the opposite direction ($N \mathrm{III}$）:
```

````python
%%render 2
F=R_x_ #kN to the right
````

```{dropdown} Show code cell outputs

Conservation of mass:

$$
Q_1 = Q_2 \\
v_1 A_1 = v_2 A_2 \\
v_1 y_1 b = v_2 y_2 b \\
v_2 = \frac{y_1 v_1}{y_2}
$$

Conservation of Energy:

$$
z_1 + \frac{P_1}{\rho g} + \frac{v_1^2}{2g} = z_2 + \frac{P_2}{\rho g} + \frac{v_2^2}{2g} + h_f \\
\frac{P_1}{\rho g} = y_1, \quad \frac{P_2}{\rho g} = y_2 \\
z_1 = z_2
$$

$$
\frac{v_1^2}{2g} = y_2 + \frac{v_2^2}{2g} + h_f - y_1 \\
\frac{v_1^2}{2g} = y_2 + \frac{y_1^2 v_1^2}{2g y_2^2} + h_f - y_1
$$

Once the velocities at both sections are determined, the conservation of momentum can then be applied:

$$
\sum F_x = \rho Q (v_2 - v_1) \\
F_1 - F_2 -R_x = \rho Q (v_2 - v_1)
$$

The centroid of the spillway is:

$$
\rho g \frac{y_1}{2} A_1 - \rho g \frac{y_2}{2} A_2 - F = \rho Q (v_2 - v_1)
$$



```

````python
%%render long 
b= 100#m 
y_1= 20 #m 
y_2= 1.5 #m 
h_f= 1.2 #m 


v_1= (((y_1-y_2-h_f)*2*g)/((y_1/y_2)**2-1))**0.5 #m/s
v_2= y_1*v_1/y_2#m/s
Q= v_1*y_1*b #m3/s
A_1= b*y_1 #m2
A_2= b*y_2 #m2
R_x= rho_water*g*y_1/2*A_1-rho_water*g*y_2/2*A_2-rho_water*Q*(v_2-v_1) #N
R_xMN=R_x/1000000 #MN
````

```{dropdown} Show code cell outputs
$$
z_1+\frac {p_1}{\rho g}+\frac{v_1^2}{2g}=z_2+\frac {p_2}{\rho g}+\frac{v_2^2}{2g}\\
$$

\begin{align*}
\text{but} \quad z_1 &= z_2 \quad;\quad \text{bend in the horizontal plane}\\
v_1 &= v_2 \quad; \quad \text{Conservation of mass}\\
\text{thus} \quad p_1 &= p_2
\end{align*}

Conservation of momentum in x:  

$$
\sum F_x = \rho Q (v_{2x}-v_{1x})\\
$$
$$
R_x-p_2A_2 = \rho Q (v_{2x}-v_{1x})\\
$$

Conservation of momentum in z:  

$$
\sum F_z = \rho Q (v_{2z}-v_{1z})\\
$$
$$
p_1A_1-R_z = \rho Q (v_{2z}-v_{1z})\\
$$

```
